Построить частотный график для двух параметров: Freedom и positive affect. Какое это распределение?

```python
n, bins, patches = plt.hist(df['Freedom'], 10, facecolor='g')

plt.xlabel('frequency')
plt.ylabel('Freedom')
plt.title('Частотный график Freedom')
plt.grid(True)
plt.show()

n, bins, patches = plt.hist(df['Positive affect'], 10, facecolor='r')

plt.xlabel('frequency')
plt.ylabel('Positive affect')
plt.title('Частотный график Positive affect')
plt.grid(True)
plt.show()
```

Распределение равномерное в обоих случаях

```python
Freedom = df['Freedom']
Positive_affect = df['Positive affect']
fig, ax = plt.subplots()
ax.scatter(Freedom, Positive_affect, c='red', alpha=0.3, edgecolors='none')

ax.grid(True)
plt.xlabel('Freedom')
plt.ylabel('Positive affect')
plt.title('Точечный график двух случайных величин')
plt.show()
```

Найти медиану, моду, математическое ожидание, стандартное отклонение,  дисперсию, минимум и максимум, найти выбросы (если есть), размах для колонки Freedom

```python
print('медиана: ', statistics.median(df['Freedom']))
try:
    print('мода: ', statistics.mode(df['Freedom']))
except:
    print('мода: no unique mode; found 140 equally common values')
print('математическое ожидание: ', statistics.mean(df['Freedom']))
print('стандартное отклонение: ', statistics.pstdev(df['Freedom']))
print('дисперсия: ', statistics.pvariance(df['Freedom']))
print('минимум: ', min(df['Freedom']))
print('максимум: ', max(df['Freedom']))
print('размах: ', max(df['Freedom']) - min(df['Freedom']))
```

Почему нельзя найти моду?

мода не существует, потому что в выборке Freedom все элементы уникальные

Посмотреть корреляцию между Freedom и positive affect (графически и математически). Графически - plt.scatter, математически - посчитать коэффициент Пирсона. О чем говорит знак коэффициента?

```python
from scipy.stats import pearsonr

corr, p_value = pearsonr(Freedom, Positive_affect)
plt.scatter(Freedom, Positive_affect, marker='.')
print('Коэффициент корреляции Пирсона:', corr)
```

Коэффициент корреляции Пирсона: 0.7
говорит о том, что Freedom и Positive_affect растут пропорционально и достаточно коррелируют м-у собой
